All non-ascii chars are em dash or chinese countries. We'll fix these shortly.

```python
em_dash_cols_period1 = list(range(220, 229)) + [264]
em_dash_cols_period2 = list(range(218, 227))

# need extra lists as pandas columns are immutable (frozen list)
idx1_cols = list(idx1.columns)
idx2_cols = list(idx2.columns)

for col_idx in em_dash_cols_period1:
    idx1_cols[col_idx] = idx1_cols[col_idx].replace("–", "-")

for col_idx in em_dash_cols_period2:
    idx2_cols[col_idx] = idx2_cols[col_idx].replace("–", "-")

idx1.columns = idx1_cols
idx2.columns = idx2_cols
```

```python
chinese_to_english = {
    "美国": "USA",
    "智利": "CHL",        # Chile
    "中国": "CHN",
    "哥伦比亚": "COL",    # Colombia
    "丹麦": "DNK",        # Denmark
    "希腊": "GRC",        # Greece
    "爱尔兰": "IRL",      # Ireland
    "日本": "JPN",
    "韩国": "KOR",        # South Korea
    "新西兰": "NZL",      # New Zealand
    "俄罗斯": "RUS",
    "财务压力": "FSI",     # TODO: Financial Stress Index - confirm country or worldwide
    "克罗地亚": "HRV",     # Croatia
    "巴基斯坦": "PAK",      
    "新加坡": "SGP"
}

idx1_cols = list(idx1.columns)
idx2_cols = list(idx2.columns)

for i, col in enumerate(idx1_cols):
    for chinese, english in chinese_to_english.items():
        if chinese in col:
            idx1_cols[i] = idx1_cols[i].replace(chinese, english)

for i, col in enumerate(idx2_cols):
    for chinese, english in chinese_to_english.items():
        if chinese in col:
            idx2_cols[i] = idx2_cols[i].replace(chinese, english)

idx1.columns = idx1_cols
idx2.columns = idx2_cols
```

```python
# log returns most symmetric
print(bdi["BDI"].skew())
print(np.log(bdi["BDI"]).skew())
print(np.log(bdi["BDI"]).diff().skew())
```

```python
from statsmodels.graphics.tsaplots import plot_acf
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.tsa.stattools import adfuller

targets = {"BDI": bdi["BDI"], "log": np.log(bdi["BDI"]), "Log return": np.log(bdi["BDI"]).diff().dropna()}

for name, series in targets.items(): 
    fig, ax = plt.subplots(1, 2)
    plot_acf(series.values, ax=ax[0])
    plot_pacf(series.values, ax=ax[1]);
    fig.suptitle(f"{name}, adf p-value = {adfuller(series.values)[1]:.3f}")
    fig.tight_layout()
    fig.show()
```

From the acf/pcf plots and adf stats, we have clear evidence that BDI/log BDI are non-stationary, while the log-returns appear to be mostly stationary. We should definitely consider predicting the log-return as an alternative target. As expected, the strongest partial auto-correlation comes from the previous value (lag 1), so the martingale model should definitely be included as naive baseline.

```python
fig, ax = plt.subplots()
for window in [3, 6, 9, 12]:
    z = (bdi["BDI"] - bdi["BDI"].rolling(window=window).mean()) / bdi["BDI"].rolling(window=window).std()
    z.dropna().plot(label=f"window = {window}", ax=ax, alpha=0.5);
    ax.legend()
fig.suptitle("Rolling-Z scores BDI");
```

```python

fig, ax = plt.subplots()
for window in [3, 6, 9, 12]:
    z = (targets["Log return"] - targets["Log return"].rolling(window=window).mean()) / targets["Log return"].rolling(window=window).std()
    z.dropna().plot(label=f"window = {window}", ax=ax, alpha=0.5);
    ax.legend()
fig.suptitle("Rolling-Z scores Log-Return");
```
